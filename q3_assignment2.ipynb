Complete the TensorFlow code. The aim is to write a model that can predict the next point in a time series using an LTSM layer with 128 units and a single dense layer in keras. You will need to fit the model in order to answer all the question

```python
import tensorflow as tf
import numpy as np
```

```python
x = np.linspace(0,242, num=242)
y_true = 10*np.sin(x/2) + 0.00002 * x**3 + -0.006 * x**2 + 0.1 * x
y = y_true + np.random.normal(scale=3, size=len(x))
```

```python
n_window = 5   # size of the sequence to use for prediction
n_input =  1   # dimension of the input data
n_output = 1   # dimension of the prediction
size_train = 201 # how much training data to use
```

```python
# Split the data into training and test data
train = x[:size_train]
test = x[size_train:]
```

```python
# reshape into batches of length given by window
X_batch = train[:-1].reshape(-1, n_window, n_input)
Y_batch = train[n_window::n_window].reshape(-1, n_input)

X_test = test[:-1].reshape(-1, n_window, n_input)
Y_test = test[n_window::n_window].reshape(-1, n_input)
```

```python
print(X_batch.shape)
print(Y_batch.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
# fit the model
model = tf.keras.Sequential()
model.add(tf.keras.layers.LSTM(units=128))       #MULTIPLE OPTIONS?
model.add(tf.keras.layers.Dense(1))
```

```python
#compile the model

model.compile(loss='mean_squared_error', optimizer='Adam', metrics=['accuracy'])
model.fit(X_batch, Y_batch, epochs=2000, batch_size=X_batch.shape[0], verbose=1)
```
