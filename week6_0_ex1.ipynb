1-dimensional convolution in numpy

```python
import matplotlib.pyplot as plt
import numpy as np


f = np.zeros((200))
f[100:]=1
fig = plt.figure(figsize=(4,3))
plt.plot(f,'-.')
xmin, xmax = plt.xlim(75,125)
plt.show()
```

```python
#Now we’ll create the kernel for the convolution
g = np.ones((21))/21.0
g
```

```python
#We’ll use the in-built numpy function, convolve to perform the convolution,
#  it will take the average of all the values of f within 10 elements either
#side of the current position of the convolution. In effect it is going to smooth out the step function we have created
f_g=np.convolve(f,g, 'same')
```

```python
fig = plt.figure(figsize=(4,3))
plt.plot(f,'-.', label='original vector')
plt.plot(f_g,label='convolution')

plt.legend()
xmin, xmax = plt.xlim(75,125)
plt.show()
```
