## Week 6 - Example 2

detect edges in an image by applying different convolution operations to the original
image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('iris.png')[...,::-1]
               
# next we plot the original image
fig, axs = plt.subplots(2,2)
                
ax0 = axs[0,0]
ax0.imshow(img)
ax0.set_title('Original')
ax0.get_xaxis().set_visible(False)
ax0.get_yaxis().set_visible(False)


# here, we'll apply the Laplacian kernel, a common edge detecting kernel
ax1 = axs[0,1]
laplacian = np.array(([0, 1, 0],[1, -4, 1],[0, 1, 0]), dtype="int")
dst = cv2.filter2D(img,-1,laplacian)
ax1.imshow(dst)
ax1.set_title('Laplacian kernel')
ax1.get_xaxis().set_visible(False)
ax1.get_yaxis().set_visible(False)

# next, we'll use a Sobel kernel to detect horizontal edges
ax2 = axs[1,0]
sobelX = np.array(([-1, 0, 1],[-2, 0, 2],[-1, 0, 1]), dtype="int")
dst = cv2.filter2D(img,-1,sobelX)
ax2.imshow(dst)
ax2.set_title('Sobel X kernel')
ax2.get_xaxis().set_visible(False)
ax2.get_yaxis().set_visible(False)


# finally, we'll use a slightly different Sobel kernel to detect vertical edges
ax3 = axs[1,1]
sobelY = np.array(([-1, -2, -1],[0, 0, 0],[1, 2, 1]), dtype="int")
dst = cv2.filter2D(img,-1,sobelY)
ax3.imshow(dst)
ax3.set_title('Sobel Y kernel')
ax3.get_xaxis().set_visible(False)
ax3.get_yaxis().set_visible(False)
plt.show()
```
