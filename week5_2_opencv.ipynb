## Drawing

```python
img = np.zeros((250,250,3), dtype=np.uint8)
img = cv2.line(img,(0,0),(100,100),(255,0,0),5)
img = cv2.circle(img,(100,100),50,(0,255,0),5)
fig = plt.figure(tight_layout=True)
im = plt.imshow(img[...,::-1])
plt.show()
```

## Basic image Operations

```python
img = cv2.imread('iris.png')
print(img.shape)
#This prints the images resolution for you. Remember that OpenCV loads the y-coordinate first in an array, so the
#resolution returned is (y, x, colour channels)
```

```python
roi = img[280:680, 330:730,:]
```

```python
#Next well take the section of image loaded into roi and pastes it into the top left corner of the image
img[0:400, 0:400,:] = roi

fig = plt.figure(tight_layout=True)
im = plt.imshow(img[...,::-1])
plt.show()
```
